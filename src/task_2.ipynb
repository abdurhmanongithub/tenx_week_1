```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import talib
from datetime import datetime
```

```python
stock_data = pd.read_csv('../data/AAPL_historical_data.csv')
```

```python
stock_data.head()
```

```python
stock_data['MACD'], stock_data['Signal'], _ = talib.MACD(stock_data['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
stock_data['RSI'] = talib.RSI(stock_data['Close'], timeperiod=14)
```

```python
stock_data['SMA50'] = talib.SMA(stock_data['Close'], timeperiod=50)
stock_data['SMA200'] = talib.SMA(stock_data['Close'], timeperiod=200)
```

```python
plt.figure(figsize=(14,8))
```

```python
plt.subplot(2, 1, 1)
plt.plot(stock_data['Close'], label='Close Price')
plt.plot(stock_data['SMA50'], label='50-Day SMA', linestyle='--')
plt.plot(stock_data['SMA200'], label='200-Day SMA', linestyle='--')
plt.title('AAPL Stock Price with SMAs')
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.plot(stock_data['MACD'], label='MACD', color='orange')
plt.plot(stock_data['Signal'], label='Signal Line', color='blue')
plt.title('AAPL MACD and Signal Line')
plt.legend()

plt.tight_layout()
plt.show()
```
